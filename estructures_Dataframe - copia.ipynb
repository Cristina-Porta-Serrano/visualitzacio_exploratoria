##*Exercice 3 S03 T01*##

*Integra les visualitzacions gràfiques, en la tasca 5, del Sprint 2.*

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.catplot(x="Month", kind="count", palette="Greens_d", data=del_flights_clean, order=del_flights_clean['Month'].value_counts().index)
```

En aquesta gràfica veiem els mesos amb major i menor número de vols.

```python
distance_companies = pd.DataFrame(del_flights_clean.groupby('UniqueCarrier')["Distance"].sum().sort_values(ascending=False).reset_index())
distance_companies
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.barplot(x="UniqueCarrier", y="Distance", palette="Greens_d", data=distance_companies) 
```

En aquesta gràfica veiem la distancia acumulada de cada companyia.

```python
dist_airtime_companies = pd.DataFrame(del_flights_clean.groupby('UniqueCarrier')["Distance","AirTime"].sum().sort_values(by="Distance", ascending=False))
dist_airtime_companies
```

```python
dist_airtime_companies.plot(kind="bar", color=['green', 'limegreen'])
```

En aquesta gràfica veiem la distància acumulada i el temps de vol acumulat solen anar en proporció. Normalment, les companyies que tenen més distància acumulada, han fet més temps de vol acumulat.

```python
airtime_companies_month = pd.DataFrame(del_flights_clean.groupby('Month')["Distance","AirTime","DepDelay","ArrDelay"].sum().sort_values(by="Distance", ascending=False))
airtime_companies_month
```

```python
airtime_companies_month.plot(kind="bar", color=['green', 'limegreen', 'lightgreen', 'mediumseagreen'])
```

En aquesta gràfica veiem els mesos amb més distància de vol, temps de vol i retards acumulats. Generalment, els mesos amb més distància acumulada de vol, també son els que tenen més temps de vol acumulat i retards acumulats, amb algun valor en exepció.
