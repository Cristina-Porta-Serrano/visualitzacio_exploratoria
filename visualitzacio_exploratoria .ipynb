```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##*Exercici 1*##

*Resumeix gràficament el data set DelayedFlights.csv. Crea almenys una visualització per:*

**Una variable categòrica (UniqueCarrier):**

```python
del_flights2 = pd.read_csv("DelayedFlights.csv")
```

```python
del_flights2
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.catplot(x="UniqueCarrier", kind="count", palette="Greens_d", data=del_flights2, order=del_flights2['UniqueCarrier'].value_counts().index)
plt.savefig('one_cat_var.png', dpi=300, bbox_inches='tight')
```

En aquesta gràfica veiem que hi ha una gran diferència entre el nº de vols de la companyia que en té més i la que en té menys. Per tant, hi ha dos extrems molt acusats i la secció mitjana de les demés dades és menys dispersa i estan més equilibrades.

**Una variable numèrica (ArrDelay):**

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.kdeplot(del_flights2['ArrDelay'], shade=True, bw_method=1, color="green")
plt.savefig('one_num_var.png', dpi=300, bbox_inches='tight')
plt.show()
```

En aquesta gràfica veiem que hi ha una densitat molt alta d'arribades amb poc retard i una densitat que tendeix a 0 d'arribades amb molt retard.

**Una variable numèrica i una categòrica (ArrDelay i UniqueCarrier):**

Creem un nou dataframe amb les columnes que volem utilitzar per a la gràfica. Fem un index nou per a que no agafi una de les columnes com a index, i a la columna d'ArrDelay posem el sumatori dels valors per a cada UniqueCarrier.

```python
more_delayed_flights_companies = pd.DataFrame(del_flights2.groupby('UniqueCarrier')["ArrDelay"].sum().sort_values(ascending=False).reset_index())
more_delayed_flights_companies
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.barplot(x="UniqueCarrier", y="ArrDelay", palette="Greens_d", data=more_delayed_flights_companies) 
plt.savefig('one_num_one_cat_var.png', dpi=300, bbox_inches='tight')
```

En aquesta gràfica segueix haven-t'hi molta diferència entre els valors extrems ja que, com que hi ha molts més vols d'una companyia que de l'altra, és normal que l'acumulat dels retards en els vols sigui molt més en la companyia que té més vols.

**Dues variables numèriques (ArrDelay i DepDelay):**

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.regplot(x=del_flights2["DepDelay"], y=del_flights2["ArrDelay"], color="green")
plt.savefig('two_num_var.png', dpi=300, bbox_inches='tight')
```

En aquesta gràfica veiem que hi ha una codependència bastant alta entre les dues variables numèriques analitzades, com es pot observar en la línia de regressió. Quan un avió surt amb un retard, normalment arriba amb un retard semblant. Veiem també que mentre menor és el retard de sortida, és més probable arribi amb més retard.

**Tres variables (ArrDelay, DepDelay i UniqueCarrier):**

Creem un nou dataframe amb les columnes que volem utilitzar per a la gràfica. Fem un index nou per a que no agafi una de les columnes com a index, i a les columnes d'ArrDelay i DepDelay posem el sumatori dels valors per a cada UniqueCarrier.

```python
more_delayed_companies = pd.DataFrame(del_flights2.groupby('UniqueCarrier')["ArrDelay","DepDelay"].sum().sort_values(by="UniqueCarrier", ascending=False).reset_index())
more_delayed_companies
```

First attempt:

```python
plt.plot('UniqueCarrier', 'DepDelay', data=more_delayed_companies, marker='o', markerfacecolor='green', markersize=12, color='green', linewidth=2)
plt.plot('UniqueCarrier', 'ArrDelay', data=more_delayed_companies, marker='+', markerfacecolor='limegreen', markersize=12, color='limegreen', linewidth=2, linestyle='dashed')

plt.legend()

plt.show()
```

En aquesta gràfica tornem a veure que hi ha una codependència bastant alta entre les dues variables numèriques analitzades, com es pot veure en la superposició de les dues línies. Quan un avió surt amb un retard, normalment arriba amb un retard semblant. Veiem també que la companyia amb major retard acumulat de sortida, sol recuperar aquest retard i arriba amb menys retard.

És cert que aquest tipus de gràfic no seria el més adequat per representar aquestes variables perquè no hi ha relació entre els punts (els Unique Carriers). Per tant, fem un altre tipus de gràfic.

Second attempt:

Creem un nou dataframe amb les columnes que volem utilitzar per a la gràfica. Aquí no fem un index nou perquè si no al gràfic apareixen els números de l'index enlloc dels UniqueCarriers, i a la columna d'ArrDelay i DepDelay posem el sumatori dels valors per a cada UniqueCarrier.

```python
more_delayed_companies2 = pd.DataFrame(del_flights2.groupby('UniqueCarrier')["ArrDelay","DepDelay"].sum().sort_values(by="DepDelay", ascending=False))
more_delayed_companies2
```

```python
more_delayed_companies2.plot(kind="bar", color=['green', 'limegreen'])
plt.savefig('three_var.png', dpi=300, bbox_inches='tight')
```

Observem el mateix que a la gràfica lineal, que els departure i arrival time solen ser molt semblant i que les companyies amb el retard en la sortida acumulat més alt, el poden recuperar a l'arribada.

**Més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier):**

First attempt:

Creem un nou dataframe amb les columnes que volem utilitzar per a la gràfica. Fem un index nou per a que no agafi una de les columnes com a index, i a les columnes d'ArrDelay, DepDelay i AirTime posem el sumatori dels valors per a cada UniqueCarrier.

```python
airtime_companies = pd.DataFrame(del_flights2.groupby('UniqueCarrier')["ArrDelay","DepDelay","AirTime"].sum().sort_values(by="UniqueCarrier", ascending=False).reset_index())
airtime_companies
```

```python
plt.plot('UniqueCarrier', 'DepDelay', data=airtime_companies, marker='o', markerfacecolor='green', markersize=12, color='green', linewidth=2)
plt.plot('UniqueCarrier', 'ArrDelay', data=airtime_companies, marker='+', markerfacecolor='limegreen', markersize=12, color='limegreen', linewidth=2, linestyle='dashed')
plt.plot('UniqueCarrier', 'AirTime', data=airtime_companies, marker='v', markerfacecolor='lightgreen', markersize=12, color='lightgreen', linewidth=2, linestyle='dotted')

plt.legend()

plt.show()
```

En aquesta gràfica veiem un creixement similar entre els retards acumulats i el temps de vol acumulat. També veiem que hi ha algunes companyies (WN i AA sobretot, i en menor mesura UA, DL i CO) que acumulen molt temps de vol i tenen pocs retards en comparació.

És cert que aquest tipus de gràfic no seria el més adequat per representar aquestes variables perquè no hi ha relació entre els punts (els Unique Carriers). Per tant, fem un altre tipus de gràfic.

Second attempt:

Creem un nou dataframe amb les columnes que volem utilitzar per a la gràfica. Aquí no fem un index nou perquè si no al gràfic apareixen els números de l'index enlloc dels UniqueCarriers, i a les columnes d'ArrDelay,DepDelay i AirTime posem el sumatori dels valors per a cada UniqueCarrier.

```python
airtime_companies2 = pd.DataFrame(del_flights2.groupby('UniqueCarrier')["ArrDelay","DepDelay","AirTime"].sum().sort_values(by="AirTime", ascending=False))
airtime_companies2
```

```python
airtime_companies2.plot(kind="bar", color=['green', 'limegreen', 'lightgreen'])
plt.savefig('four_var.png', dpi=300, bbox_inches='tight')
```

Observem el mateix que a la gràfica lineal, un creixement similar entre els retards acumulats i el temps de vol acumulat. També veiem que hi ha algunes companyies (WN i AA sobretot, i en menor mesura UA, DL i CO) que acumulen molt temps de vol i tenen pocs retards en comparació.
